```python
from decimal import Decimal
import matplotlib
import matplotlib.pyplot as plt
from ipywidgets import widgets
import math
import pylab
'''
Methods
'''


def common_part(define_y, x1, x2, h, x0, y0):
    """
    Common part for all methods, used only inside the methods
    """
    lx = list()
    ly = list()
    n = int((x2 - x1) / (h)) + 1
    last_y = y0
    last_x = x0
    for i in range(n):
        lx.append(float(last_x))
        ly.append(float(last_y))
        y = define_y(last_x, last_y)
        last_y = y
        last_x += h
    return lx, ly

def Euler_method(f, x1, x2, h, x0, y0):
    """
    Euler method
    f: DE
    x1, x2: xE[x1 ,x2]
    h: step
    x0: f(x0) = y0
    y0: f(x0) = y0
    return: lx: list of x
            ly: list of y
            where x and y are found at every step according to the given segment
    """
    def define_y(last_x, last_y):
        return last_y + h * f(last_x, last_y)


    return common_part(define_y, x1, x2, h, x0, y0)

    
def improved_Euler_method(f, x1, x2, h, x0, y0):
    """
    Improved Euler method
    f: DE
    x1, x2: xE[x1 ,x2]
    h: step
    x0: f(x0) = y0
    y0: f(x0) = y0
    return: lx: list of x
            ly: list of y
            where x and y are found at every step according to the given segment
    """
    def define_delta_y(last_x, last_y):
        return h * f(last_x + h / 2, last_y + h / 2 * f(last_x, last_y))

    def define_y(last_x, last_y):
        return last_y + define_delta_y(last_x, last_y)

    return common_part(define_y, x1, x2, h, x0, y0)


def Runge_Kutta_method(f, x1, x2, h, x0, y0):
    """
    Runge_Kutta method
    f: DE
    x1, x2: xE[x1 ,x2]
    h: step
    x0: f(x0) = y0
    y0: f(x0) = y0
    return: lx: list of x
            ly: list of y
            where x and y are found at every step according to the given segment
    """
    def define_k1(last_x, last_y):
        return f(last_x, last_y)

    def define_k2(last_x, last_y, k1):
        return f(last_x + h / 2, last_y + (h * k1 / 2))

    def define_k3(last_x, last_y, k2):
        return f(last_x + h / 2, last_y + (h * k2 / 2))

    def define_k4(last_x, last_y, k3):
        return f(last_x + h, last_y + h * k3)

    def define_delta_y(last_x, last_y):
        k1 = define_k1(last_x, last_y)
        k2 = define_k2(last_x, last_y, k1)
        k3 = define_k3(last_x, last_y, k2)
        k4 = define_k4(last_x, last_y, k3)
        return h / 6 * (k1 + 2 * k2 + 2 * k3 + k4)

    def define_y(last_x, last_y):
        return last_y + define_delta_y(last_x, last_y)

    return common_part(define_y, x1, x2, h, x0, y0)
```

```python
'''
Code implementing the peculiar properties of differential equation
'''
def my_DE(x, y):
    """
    Differential equation, according to my number from group
    """
    return Decimal(math.sin(x)) + Decimal(y)

def exact_solution(x, c1):
    """
    Solution to my differential equation
    x: f(x)=y
    c1: constant
    """    
    return c1 * Decimal(math.exp(x)) - Decimal(math.sin(x))/2 - Decimal(math.cos(x))/2

def define_c1(x0, y0):
    """
    Function defining constant
    x0: f(x0)=y0
    y0: f(x0)=y0
    return constant
    """
    return (Decimal(y0) + Decimal(math.sin(x0)/2) + Decimal(math.cos(x0)/2)) / Decimal(math.exp(x0))

def find_points_of_exact_solution(define_y, x1, x2, h, x0, y0):
    """
    define_y: exact solution function
    x1, x2: xE[x1 ,x2]
    h: step
    x0: f(x0) = y0
    y0: f(x0) = y0    
    return: lx: list of x
            ly: list of y
            where x and y are found at every step according to the given segment
    """
    c1 = define_c1(x0, y0)
    print(c1)
    lx = list()
    ly = list()
    n = int((x2 - x1) / (h)) + 1
    last_y = y0
    last_x = x0
    for i in range(n):
        lx.append(Decimal(last_x))
        ly.append(Decimal(last_y))
        last_x += h
        y = define_y(last_x, c1)
        last_y = y
    return lx, ly


```

```python
'''
Inputs, you can change inputs there if you want
'''
y0 = Decimal('1')  # f(x0)=y0
x0 = Decimal('0') # f(x0)=y0
h = Decimal('0.1') # step, affects precision
x2 = Decimal('12')/Decimal('5') #end of interval
x1 = Decimal(x0) #start of interval
```

```python
'''
Find points according to function and methods
'''
Euler_lx, Euler_ly = Euler_method(my_DE, x1, x2, h, x0, y0) #Euler results in lists 
imp_Euler_lx, imp_Euler_ly = improved_Euler_method(my_DE, x1, x2, h, x0, y0) #Improved Euler results lists
Runge_Kutta_lx, Runge_Kutta_ly = Runge_Kutta_method(my_DE, x1, x2, h, x0, y0) # Runge-Kutta results in list

exact_lx, exact_ly = find_points_of_exact_solution(exact_solution, x1, x2, h, x0, y0) #Exact solution results in lists
```

```python
'''
Code for graphs
'''
#Figure for all graphs
fig = plt.figure(figsize=(90, 40))

#Euler method
sub = fig.add_subplot(1, 3, 1)
plt.plot(Euler_lx, Euler_ly, label = 'Euler method') 
plt.plot(exact_lx, exact_ly, label = 'exact') 
sub.legend(loc='upper center', bbox_to_anchor=(0.5, 1.00), shadow=True, ncol=3, prop={'size': 50})

#Improved Euler method
sub = fig.add_subplot(1, 3, 2)
plt.plot(imp_Euler_lx, imp_Euler_ly, label = 'improved Euler method') 
plt.plot(exact_lx, exact_ly, label = 'exact') 
sub.legend(loc='upper center', bbox_to_anchor=(0.5, 1.00), shadow=True, ncol=3, prop={'size': 50})

#Runge-Kutta method
sub = fig.add_subplot(1, 3, 3)
plt.plot(Runge_Kutta_lx, Runge_Kutta_ly, label = 'Runge-Kutta method') 
plt.plot(exact_lx, exact_ly, label = 'exact') 
sub.legend(loc='upper center', bbox_to_anchor=(0.5, 1.00), shadow=True, ncol=3, prop={'size': 50})

plt.show()
#That takes time because of size of plots
```

```python
"""Find local and global errors"""

#Lists for local and global errors
local_errors_Euler = list()
local_errors_imp_Euler = list()
local_errors_Runge_Kutta = list()
global_error_Euler = 0
global_error_imp_Euler = 0
global_error_Runge_Kutta = 0

def find_local_errors():
    """
    Function that finds local errors
    """
    for i in range(len(exact_lx)):
        local_errors_Euler.append(Decimal(abs((Decimal(exact_ly[i]))-Decimal(Euler_ly[i]))))
        local_errors_imp_Euler.append(Decimal(abs(Decimal(exact_ly[i])-Decimal(imp_Euler_ly[i]))))
        local_errors_Runge_Kutta.append(Decimal(abs(Decimal(exact_ly[i])-Decimal(Runge_Kutta_ly[i]))))
def find_global_errors():
    """
    Function that finds global errors
    """
    global global_error_Euler
    global global_error_imp_Euler
    global global_error_Runge_Kutta
    for i in range(len(exact_ly)):
        if local_errors_Euler[i]>global_error_Euler:
            global_error_Euler = local_errors_Euler[i]
        if local_errors_imp_Euler[i]>global_error_imp_Euler:
            global_error_imp_Euler = local_errors_imp_Euler[i]
        if local_errors_Runge_Kutta[i]> global_error_Runge_Kutta:
            global_error_Runge_Kutta = local_errors_Runge_Kutta[i]
#I know that is must be the last local error, because it is growing,
#according to Shilov, but I did it as people from group said    
find_local_errors()
find_global_errors()
        
        
        
```

```python
'''
Code implementing tables
'''

def print_matrix(list_of_list):
    """
    Function that implements table
    """
    number_width = ((max([max(len(x) for x in i) for i in list_of_list])))
    cols = max(map(len, list_of_list))
    output = '+'+('-'*(number_width+2)+'+')*cols + '\n'
    for row in list_of_list:
        for column in row:
            output += '|' + ' {:^{width}s} '.format(column, width = number_width)
        output+='|\n+'+('-'*(number_width+2)+'+')*cols + '\n'
    return output

def create_list_of_list(list_x, list_y, list_local_errors):
    """
    Make data approvable by print_matrix function
    """
    data = [["step","x","y","local error"]]
    for i in range(len(exact_lx)): 
        temp = list()
        temp.append(str(i))
        #Left only 4 digits after the dot in x values
        temp.append(f'{list_x[i]:.04f}')
        #Left only 4 digits after the dot in y values
        temp.append(f'{list_y[i]:.04f}')
        #Left only 10 digits after the dot in local errors
        temp.append(f'{list_local_errors[i]:.10f}')
        data.append(temp)
    return data

#Print tables
print("Euler method")
print('Global error: ',global_error_Euler )
print(print_matrix(create_list_of_list(Euler_lx, Euler_ly, local_errors_Euler)))

print("Improved Euler method")
print('Global error: ',global_error_imp_Euler )
print(print_matrix(create_list_of_list(imp_Euler_lx, imp_Euler_ly, local_errors_imp_Euler)))

print("Runge-Kutta method")
print('Global error: ',global_error_Runge_Kutta )
print(print_matrix(create_list_of_list(Runge_Kutta_lx, Runge_Kutta_ly, local_errors_Runge_Kutta)))
 
```
